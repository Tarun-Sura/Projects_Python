# Movie List of top 250 movies in IMDB

```python
import requests
import re
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = 'http://www.imdb.com/chart/top'
def Extract_link(link):
    
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'lxml')
    d = {'name':[],'rate':[],'link':[]}
    for a,rate in zip(soup.select('td.titleColumn'),soup.select('td.ratingColumn strong')):
        name = a.find('a').text
        link = a.find('a').get('href')
        rate = rate.text
        #print(f"Movie Name:{name}  ,Rating:{rate}")
        #print(f"Link here: https://www.imdb.com{link}")
        links = (f"https://www.imdb.com{link}")
        d['name'].append(name)
        d['rate'].append(rate)
        d['link'].append(links)
        
    top_list_df = pd.DataFrame(d)    
    return top_list_df  

top_list_df = Extract_link(url)  
top_list_df.head(20)
```

## Exporting to CSV


```python

top_list_df.to_csv('movie_top250.csv',columns=['name','rate','link'])
```
